## 6.5 __이미지 분류__

* cifar-10 데이터를 가져와 훈련데이터와 평가데이터로 분리

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
X_train.size/(32*32*3) # 최소 50,000개의 데이터를 훈련데이터로 분리
```

```python
X_test.size/(32*32*3) # 다음 10,000개의 데이터를 평가데이터로 분리
```

* 이미지 데이터의 구조와 표현

__분류작업__ (1) 데이터 전처리 : RGB 픽셀이미지를 1차원 벡터로 변환

```python
X_train_flat = X_train.reshape(X_train.shape[0], -1)
y_train_flat = y_train.reshape(-1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)
y_test_flat = y_test.reshape(-1)
# 필요하다고 판단되면 0과 1사이의 값으로 정규화 과정 추가
```

__분류작업__ (2) 로지스틱 회귀모형을 이용한 분류모형 학습

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
LR = LogisticRegression(max_iter=100,n_jobs=-1)
LR.fit(X_train_flat,y_train_flat)
```

```python
LR.score(X_train_flat,y_train_flat)
```

```python
LR.score(X_test_flat,y_test_flat)
```

```python
y_test_pred = LR.predict(X_test_flat)
y_train_pred = LR.predict(X_train_flat)
print(classification_report(y_train_flat, y_train_pred))
print(classification_report(y_test_flat, y_test_pred))
```

__분류작업__ (3) kNN모형을 이용한 분류모형 학습

```python
from sklearn.neighbors import KNeighborsClassifier
KC = KNeighborsClassifier(n_jobs=-1)
KC.fit(X_train_flat,y_train_flat)
```

```python
KC.score(X_train_flat,y_train_flat)
```

```python
KC.score(X_test_flat,y_test_flat)
```

```python
y_test_pred = KC.predict(X_test_flat)
y_train_pred = KC.predict(X_train_flat)
print(classification_report(y_train_flat, y_train_pred))
print(classification_report(y_test_flat, y_test_pred))
```

__분류작업__ (4) 분류나무모형을 이용한 분류모형 학습

```python
from sklearn.tree import DecisionTreeClassifier
DC = DecisionTreeClassifier(random_state=3)
DC.fit(X_train_flat,y_train_flat)
```

```python
DC.score(X_train_flat,y_train_flat)
```

```python
DC.score(X_test_flat,y_test_flat)
```

```python
y_test_pred = DC.predict(X_test_flat)
y_train_pred = DC.predict(X_train_flat)
print(classification_report(y_train_flat, y_train_pred))
print(classification_report(y_test_flat, y_test_pred))
```
