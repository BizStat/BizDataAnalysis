(1) 파이썬으로 작성한 간단한 계산기 프로그램

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero!"
    return x / y

print("간단한 계산기 프로그램입니다.")
print("사칙 연산을 선택하세요:")
print("1. 더하기")
print("2. 빼기")
print("3. 곱하기")
print("4. 나누기")

while True:
    choice = input("원하는 연산을 선택하세요 (1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        num1 = float(input("첫 번째 숫자를 입력하세요: "))
        num2 = float(input("두 번째 숫자를 입력하세요: "))

        if choice == '1':
            print("결과:", add(num1, num2))
        elif choice == '2':
            print("결과:", subtract(num1, num2))
        elif choice == '3':
            print("결과:", multiply(num1, num2))
        elif choice == '4':
            print("결과:", divide(num1, num2))

        break
    else:
        print("올바른 연산자를 선택하세요.")
```

(2) 계산기 프로그램을 GUI 형태로 수정

```python
import tkinter as tk

def calculate():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        operation = operation_var.get()

        if operation == "+":
            result = num1 + num2
        elif operation == "-":
            result = num1 - num2
        elif operation == "*":
            result = num1 * num2
        elif operation == "/":
            if num2 == 0:
                result = "Error: Division by zero!"
            else:
                result = num1 / num2

        result_label.config(text="Result: " + str(result))
    except ValueError:
        result_label.config(text="Error: Invalid input")

# GUI 생성
root = tk.Tk()
root.title("간단한 계산기")

# 두 개의 숫자 입력 필드
entry1 = tk.Entry(root)
entry1.pack(pady=5)
entry2 = tk.Entry(root)
entry2.pack(pady=5)

# 연산 선택 옵션 메뉴
operation_var = tk.StringVar(root)
operation_var.set("+")  # 초기값 설정
operation_menu = tk.OptionMenu(root, operation_var, "+", "-", "*", "/")
operation_menu.pack(pady=5)

# 계산 버튼
calculate_button = tk.Button(root, text="Calculate", command=calculate)
calculate_button.pack(pady=5)

# 결과 표시 레이블
result_label = tk.Label(root, text="Result: ")
result_label.pack(pady=5)

root.mainloop()
```

* 설치된 한글 글꼴을 지정하여 한글 워드 클라우드를 작성한 사례

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 나눔 폰트 경로 설정
font_path = "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf"

# 텍스트 데이터 예시
text = "경기대학교 경영학전공은 마케팅, 생산관리, 인사조직, 재무, 회계, 국제경영 \
등 경영학 전반에 걸쳐 교육하고 있으며, 지금까지 경영학전공을 거쳐 간 졸업생들은 \
사회각계각층에서 큰 활약을 하고 있습니다. 경영학전공은 21세기를 대표하는 경영인을 \
양성하기 위하여 창의적이고 진취적인 기업가 정신 및 전문 직업의식의 고취, 올바른 \
기업윤리 정신과 건전한 함양, 전문적이고 실용적인 경영학 지식, 어학능력을 배양하고자 \
노력하고 있습니다"

# 워드 클라우드 생성
wordcloud = WordCloud(width=800, height=400, background_color='white',font_path=font_path).generate(text)

# 워드 클라우드 이미지 표시
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# 워드 클라우드 이미지 파일로 저장
wordcloud.to_file("wordcloud_example.png")
```
