## 5.2 __기계학습 사례 실습__

```python
import pandas as pd
```

* 실습 데이터 가져오기

```python
income = pd.read_table('/content/BizDataAnalysis/DATA/Income.csv',sep=',')
```

```python
income.head()
```

* 데이터 분리

```python
train = income.iloc[:20]
test = income.iloc[21:]
```

```python
test.reset_index()
```

* 선형회귀모형

```python
def lin_func(x,b0,b1):
  y = b0+b1*x
  return y
```

```python
from scipy.optimize import curve_fit
betas, covs = curve_fit(lin_func, train['Education'], train['Income'])
```

```python
betas
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.plot(train['Education'], train['Income'], 'bo', label='data')
plt.plot(train['Education'],betas[0]+betas[1]*train['Education'],color='red')
```

* 모형의 평가

```python
test['linfit'] = lin_func(test['Education'],betas[0],betas[1])
```

```python
test
```

```python
plt.figure(figsize=(8,8))
plt.plot(test['Education'], test['Income'], 'bo', label='data')
plt.plot(test['Education'],test['linfit'],color='red')
```

```python
import numpy as np
np.sqrt(sum((test['Income']-test['linfit'])**2)/len(test))
```

```python
np.sqrt(sum(np.abs(test['Income']-test['linfit']))/len(test))
```

* 비선형회귀모형

```python
def nonlin_func(x,t1,t2,t3,t4):
  y = t1 + t2*x + t3*(x**2) + t4*(x**3)
#  y = t1 + (t2 - t1) * np.exp(-t3 * (x**t4))
  return y
```

```python
thetas, covs = curve_fit(nonlin_func, train['Education'], train['Income'])
```

```python
thetas
```

```python
train['nonlinfit'] = nonlin_func(train['Education'],thetas[0],thetas[1],thetas[2],thetas[3])
```

```python
train.sort_values('nonlinfit')
```

```python
plt.figure(figsize=(8,8))
plt.plot(train['Education'], train['Income'], 'bo', label='observed value')
plt.plot(train['Education'],train['nonlinfit'],'ro', label='fitted value')
plt.legend()
plt.show()
```

```python
test['nonlinfit'] = nonlin_func(test['Education'],thetas[0],thetas[1],thetas[2],thetas[3])
```

```python
np.sqrt(sum((test['Income']-test['nonlinfit'])**2)/len(test))
```

```python
np.sqrt(sum(np.abs(test['Income']-test['nonlinfit']))/len(test))
```
