(1) 다양한 기술통계량 산출

```python
seoul_air = pd.read_csv('/content/BizDataAnalysis/DATA/AIR_HOUR_2022.csv')
seoul_air['측정일시'] = seoul_air['측정일시'].astype('str')
seoul_air['측정일시'] = seoul_air['측정일시'].apply(lambda x : dt.datetime.strptime(x,'%Y%m%d%H'))
seoul_air['요일'] = seoul_air['측정일시'].dt.day_name()
seoul_air['시간'] = seoul_air['측정일시'].dt.hour
측정소정보 = pd.read_csv('/content/BizDataAnalysis/DATA/서울시 대기오염 측정소 정보.csv',encoding='EUC-KR')
측정항목정보 = pd.read_csv('/content/BizDataAnalysis/DATA/서울시 대기오염 측정항목 정보.csv',encoding='EUC-KR')
seoul_air = pd.merge(seoul_air,측정소정보,how='left',on='측정소 코드')
seoul_air = pd.merge(seoul_air,측정항목정보,how='left',left_on='측정항목', right_on='측정항목 코드')
seoul_air.rename(columns={'측정항목 명(줄임 명칭)' : '측정항목 이름'},inplace=True)
tmp01 = seoul_air[['요일','시간','평균값','측정기 상태','측정소 이름','측정항목 이름']]
tot = tmp01[tmp01['측정기 상태'] == 0]
```

```python
# PM10에 대한 기술통계량
stat.mean(tot['평균값'])
```

```python
tot['평균값'].mean()
```

```python
stat.mean(tot[tot['측정항목 이름'] == 'PM10']['평균값'])
```

```python
stat.stdev(tot[tot['측정항목 이름'] == 'PM10']['평균값'])
```

```python
tot[tot['측정항목 이름'] == 'PM10']['평균값'].std()
```

```python
tot[tot['측정항목 이름'] == 'SO2']['평균값'].std()
```

```python
tot[tot['측정항목 이름'] == 'PM10']['평균값'].std() / tot[tot['측정항목 이름'] == 'PM10']['평균값'].mean()
```

```python
tot[tot['측정항목 이름'] == 'SO2']['평균값'].std() / tot[tot['측정항목 이름'] == 'SO2']['평균값'].mean()
```

(2) 상자도표

```python
sns.catplot(data=tot[tot['측정항목 이름'] == 'PM10'],y='평균값',kind='box')
```

```python
sns.catplot(data=tot[tot['측정항목 이름']=='PM10'],x='요일',y='평균값',kind='box')
```

```python
sns.catplot(data=tot[tot['측정항목 이름'] == 'PM2.5'],x='요일',y='평균값',hue='측정항목 이름',kind='box')
```
