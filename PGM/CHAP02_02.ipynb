| 고객번호 | 고객이름 | 판매액(단위:원) |
|---|---|---|
|1 | 김영희 | 30000 |

```python
x1 = [1, '김영희', 30000]
x2 = ['', '',12000]
x3 = [3, '이수철', 98000]
col_name = ['고객번호','고객이름','판매액(단위:원)']
df1 = pd.DataFrame(data=[x1,x2,x3], columns=col_name)
```

```python
df1.head()
```

```python
# @title 고객이름 vs 판매액(단위:원)

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df1['고객이름'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df1, x='판매액(단위:원)', y='고객이름', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# prompt: /content/BizDataAnalysis/DATA/서울지역_일일_최대강수량.csv 를 이용해 일일 강수량이 300mm 혹은 350mm 라면 몇 년 만에 이러한 현상이 발생했는가?

df = pd.read_csv('/content/BizDataAnalysis/DATA/서울지역_일일_최대강수량.csv')

# 강수량이 300mm 또는 350mm인 날짜 찾기
target_dates = df[df['일일 최대 강수량(mm)'].isin([300, 350])]['날짜']

if not target_dates.empty:
  # 첫 번째 발생 날짜와 마지막 발생 날짜 계산
  first_date = pd.to_datetime(target_dates.iloc[0])
  last_date = pd.to_datetime(target_dates.iloc[-1])

  # 몇 년 만에 발생했는지 계산
  years_passed = (last_date - first_date).days / 365

  print(f"강수량이 300mm 또는 350mm인 현상은 {years_passed:.2f}년 만에 발생했습니다.")
else:
  print("강수량이 300mm 또는 350mm인 날짜가 없습니다.")
```
